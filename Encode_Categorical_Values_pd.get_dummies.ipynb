# Dummies for UnOrderedCategorical(Nominal) Variable not for Ordinal Variable

```python
# Dummy Variable is also know as indicator Variable.
# Dummy Variable trap helps us to remove the colinearity between features.
# 
```

```python
# import library 
import pandas as pd
```

```python
# Load the Data
train = pd.read_csv('/home/genius/Coding-My-Life/Encode_Categorical_Value-master/test_titanic.csv')
```

```python
train.head()
```

```python
# Dummies for the Series with Dictonary, contain both column
# One Way to do it- Use Series Map Method
train['Sex_male'] = train.Sex.map({'female':0, 'male':1})
train.head()
```

```python
# Other way to do this- Use of Top level Function - pd.get_dummies and it takes the train.Sex series
# Note If we have K values for categorical variable we use K-1 dummies variable to represent it
pd.get_dummies(train.Sex).iloc[:, 1:][0:12]
```

```python
# How to add this column back to the dataFrame
pd.get_dummies(train.Sex, prefix = 'Sex').iloc[:, 1:][0:12]
```

```python
# Dummies variable for the more than 2 categorical values
train.Embarked.value_counts()
```

```python
pd.get_dummies(train.Embarked, prefix = "Embarked")[0:12]
```

```python
# Dummy Variable trap 
pd.get_dummies(train.Embarked, prefix = 'Embarked').iloc[:, 1:][0:12]
```

```python
# Save to dataFrame Series
embarked_dummies = pd.get_dummies(train.Embarked, prefix = "Embarked").iloc[:, 1:]
```

```python
# Cocatenate the above series to the original dataframe
# we use concate function 0 = rows, 1 = column
train = pd.concat([train, embarked_dummies], axis = 1)
```

```python
train.head()
```

```python
# Dummies on DataFrame, drop_first feature is in pandas0.18
pd.get_dummies(train, columns = ['Sex', 'Embarked'], drop_first = True)[0:12]
```
